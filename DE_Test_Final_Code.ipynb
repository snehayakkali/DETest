```python
# Data Engineering Project Test Code

# To avoid Warning Messages
import warnings
warnings.filterwarnings('ignore')

# Importing Pandas Library
import pandas as pd

#Reading Json datafile into Python Data Frame object
df=pd.read_json('source_event_data.json',lines=True)
df.head()
```

```python
#Chekcing Size of the Data Frame
print(df.shape)
```

```python
# Chekcing details of Attributes to understand the Data
print(df.info())
```

```python
#Flattening Nested Json Data as Data Frame
df1= pd.read_json("source_event_data.json", lines=True)
df2 = pd.DataFrame(df1.user.values.tolist())
df2.head()
```

```python
# Concatenating Json and Nested Json attributes into a single Data Frame
df=pd.concat([df1['event_id'],df2['session_id'],df2['id'],df2['ip'],df1['action'],df1['url'],df1['timestamp']],axis=1)
df.head()
```

```python
# Checking Details of the Data Frame with all attributes from Json
df.info()
```

```python
#Checking for unique values of url's
df['url'].value_counts()
```

```python
# Manipulating/Cleaning url attribute data for consistency 
df['url'] = df['url'].str.strip().replace(dict(zip(['https://','www.','http://','w3.'], ['']*4)),regex=True)

df['url'].value_counts()
```

```python
# Formatting url Data 
df['url'] = 'https://www.' + df['url'].astype(str)
#df.head()
```

```python
# Validating the changes made to url attribute Data
df['url'].value_counts()
```

```python
# keeping backup of the original data before transformations
df_copy=df
df_copy.head()
```

```python
# Importing required libraries & packages
from urllib.parse import urlparse

# Adding New Columns for url Doamin and paths
df.insert(6,'url_level1','NA')
df.insert(7,'url_path','NA')

# Parsing url
for i in range(0,len(df)):
    p = urlparse(df['url'][i])
    df['url_level1'][i]=p.netloc
    df['url_path'][i]=p.path
df.head(50)

```

```python
# Separting url path data
path_split=df['url_path'].str.split("/",n=3,expand=True)
df.insert(7,'url_level2',path_split[1])
df.insert(8,'url_level3',path_split[2])

df.head()
```

```python
# Dropping unnecessary columns
df=df.drop(['url','url_path'], 1)
df.head()
```

```python
#Checking the information of Data Frame after changes
print(df.info())
```

```python
#Creating a subset of Data Frame with required columns and with change in the column names

user= df[['id','timestamp','url_level1','url_level2','url_level3','action']]
user.columns=['user_id','time_stamp','url_level1','url_level2','url_level3','activity']
user.head()
```

```python
#Chekcing the information of the subset Data Frame
user.info()
```

```python
#Writing the data to a CVS file 
user.to_csv(r'user.csv')
```

```python
# Getting value counts of the activity performed in considered timeframe for the url and paths
action_counts=pd.DataFrame(activity_count.groupby(["timestamp","url_level1","url_level2","action"])['event_id'].
                           size().reset_index(name="action_Counts"))

action_counts
```

```python
# Getting unique value counts of users for the activity performed in considered timeframe for the url Domain and paths
action_counts1=pd.DataFrame(activity_count.groupby(["timestamp","url_level1","url_level2","action"])['id'].
                        size().reset_index(name="user_Counts"))

action_counts1
```

```python
#Merging the data for user counts and action counts for each avtivity in considered frame time and url domain and paths
counts = pd.merge(action_counts, action_counts1, on=["timestamp","url_level1","url_level2","action"])

counts
```

```python
#Writing the data to a CSV file
counts.to_csv(r'Activity_User.csv')
```
